```python
df['Description']=df1['Descriptoin']
df.head()
```

```python
#to convert string type RAM & Storage columns into float values of GB,MB,TB.
import re

def convert_to_gb(value):
    if pd.isna(value):
        return None

    value = value.upper().replace(" ", "")
    
    number = float(re.findall(r"[\d.]+", value)[0])
    
    if "TB" in value:
        return number * 1024
    elif "MB" in value:
        return number / 1024
    else:  # GB
        return number
```

```python
df['RAM_GB'] = df['RAM'].apply(convert_to_gb)
df['Storage_GB'] = df['Storage'].apply(convert_to_gb)
```

```python
df.drop(columns=['RAM','Storage'],inplace=True)
df
```

```python
df.fillna(value=0,inplace=True)
df
```

```python
#Text Column which combined textual information of each product.
df['Text']=(df['Product Name']+" "+df['Description'])
```

```python
#to convert Text column into a vector of embeddings for each product.
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
text_embeddings = model.encode(
    df['Text'].tolist(),
    show_progress_bar=True
)
```

```python
df.head()
```

```python
#to convert categorical columns color and Category to one hot encodings.
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
cat_features = encoder.fit_transform(
    df[['Category', 'Color']]
)
```

```python
#to convert string type price column into float type prices.
import re

def clean_price(value):
    if pd.isna(value):
        return None
    value = str(value)
    value = re.sub(r'[^\d.]', '', value)  # keep digits & dot
    return float(value) if value else None
```

```python
df['Price'] = df['Price'].apply(clean_price)
```

```python
df.tail()
```

```python
#to scale values of numerical columns like price, RAM, Storage amd Rating to have uniform values within range 0-1.
from sklearn.preprocessing import MinMaxScaler

num_cols = ['RAM_GB', 'Storage_GB', 'Price', 'Rating']
scaler = MinMaxScaler()
num_features = scaler.fit_transform(df[num_cols])
```

```python
final_features = np.hstack([
    text_embeddings * 2.0,   
    cat_features * 1.0,      
    num_features * 0.5 ]
)  
```
